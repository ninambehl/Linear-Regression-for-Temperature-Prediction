# Model Training

## 1. Filling in Missing Values

```python
weather_history.isnull().sum()
```

```python
weather_history['Precip Type'].unique()
```

```python
weather_history['Precip Type'] = weather_history['Precip Type'].fillna("None")
```

```python
weather_history['Precip Type'].value_counts()
```

## 2. Encode

```python
weather_history.head(10)
```

### One-Hot Encoding

```python
precip_encoded = pd.get_dummies(weather_history['Precip Type'], drop_first = True, dtype = int)
```

```python
precip_encoded
```

```python
weather_history = pd.concat([weather_history, precip_encoded], axis = 1)
```

```python
weather_history.head()
```

### Label Encoding

```python
! pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encode = LabelEncoder()
```

```python
weather_history['Summary'].unique()
```

```python
weather_history['Summary Int'] = label_encode.fit_transform(weather_history['Summary'])
```

```python
weather_history.head(10)
```

## 3. Splitting Features and Labels

```python
weather_history.head(10)
```

```python
weather_history.columns
```

```python
X = weather_history[['Humidity', 'Wind Bearing (degrees)', 'rain', 'snow', 'Summary Int']]
y = weather_history['Temperature (C)']
```

### Train, Test, Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

## Training

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

### Making Predictions

```python
predictions = linear_model.predict(X_test)
```

```python
predictions
```

```python
predictions_df = pd.DataFrame({'Actual': y_test, 'Prediction': predictions})
```

```python
predictions_df
```

### Evaluation Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(y_test, predictions)
```

```python
mse
```

```python
mae = mean_absolute_error(y_test, predictions)
```

```python
mae
```
